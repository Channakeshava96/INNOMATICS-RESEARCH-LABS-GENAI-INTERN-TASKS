# Problem 1: Unique Words in a Sentence

```python
# Take input sentence from user
sentence = input("Enter sentence: ")

# Split sentence into words
words = sentence.split()

# Convert list of words into a set to remove duplicates
unique_words = set(words)

# Print count of unique words
print("Unique words count:", len(unique_words))

# Print unique words
print("Unique words:", unique_words)
```

# Problem 2: Highest Salary from Employee Data


```python
# Store employee names and their salaries in a dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find the employee with the highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

# Print the result
print("Highest Salary:", highest_employee, "-", highest_salary)
```

# Problem 3: Find Maximum and Minimum Values


```python
# List of numbers
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with the first element
max_value = numbers[0]
min_value = numbers[0]

# Loop through the list
for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

# Print results
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

# Problem 4: Count Products Above a Price Threshold


```python
# List of product prices
prices = [450, 1200, 899, 1500, 300]

# Initialize counter
count = 0

# Loop through prices
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

# Problem 5: Calculate Attendance Percentage


```python
# Attendance list
attendance = ["P", "P", "A", "P", "P"]

# Count number of present days
present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

# Total number of days
total_days = len(attendance)

# Calculate attendance percentage
attendance_percentage = (present_count / total_days) * 100

# Print result
print("Attendance Percentage:", attendance_percentage)
```

# Problem 7: Count Character Frequency


```python
# Input string
text = "python"

# Dictionary to store character frequency
char_count = {}

# Loop through each character in the string
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

# Print the result
print(char_count)
```

# Problem 9: Check if a Key Exists in a Dictionary


```python
# Dictionary of employees
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to check
key_to_check = "Ravi"

# Check if key exists
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

# Problem 10: Calculate Average Marks


```python
# List of student marks
marks = [70, 75, 80, 65]

# Calculate total marks
total_marks = sum(marks)

# Number of subjects
number_of_subjects = len(marks)

# Calculate average marks
average_marks = total_marks / number_of_subjects

# Print result
print("Average Marks:", average_marks)
```
