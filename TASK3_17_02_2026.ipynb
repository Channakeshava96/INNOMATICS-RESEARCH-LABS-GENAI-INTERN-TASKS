## Problem Statement 1: Employee Performance Bonus Eligibility

A company evaluates employee performance scores at the end of the year.
The objective is to identify the highest performance score and list all employees
who are eligible for the top performance bonus in case of a tie.


```python
# Dictionary containing employee names and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest performance score
highest_score = max(employees.values())

# List to store top performers
top_performers = []

# Identify all employees with the highest score
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

## Problem Statement 2: Search Query Keyword Analysis

This problem analyzes customer search queries by converting text to lowercase,
removing punctuation, counting keyword frequency, and displaying keywords
that appear more than once.


```python
# Input search query
query = "Buy mobile phone buy phone online"

# Convert query to lowercase
query = query.lower()

# Remove punctuation (if any)
for char in [",", ".", "!", "?", ";", ":"]:
    query = query.replace(char, "")

# Split query into words
words = query.split()

# Dictionary to store keyword frequency
keyword_count = {}

# Count frequency of each word
for word in words:
    if word in keyword_count:
        keyword_count[word] += 1
    else:
        keyword_count[word] = 1

# Store keywords searched more than once
repeated_keywords = {}

for word, count in keyword_count.items():
    if count > 1:
        repeated_keywords[word] = count

# Display result
print(repeated_keywords)
```

## Problem Statement 3: Sensor Data Validation

This problem validates hourly sensor readings by identifying even-numbered
(reading values) as valid and storing them along with their hour index.
Odd readings are ignored.


```python
# List of sensor readings (index represents hour)
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid sensor readings as (hour, value)
valid_readings = []

# Loop through readings with index
for hour, reading in enumerate(sensor_readings):
    if reading % 2 == 0:   # Check for even reading
        valid_readings.append((hour, reading))

# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## Problem Statement 4: Email Domain Usage Analysis

This problem analyzes employee email IDs to determine how many users belong
to each email domain and calculates the percentage usage of each domain.


```python
# List of employee email IDs
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store domain counts
domain_count = {}

# Extract domain and count occurrences
for email in emails:
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Total number of emails
total_emails = len(emails)

# Display percentage usage of each domain
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

## Problem Statement 5: Sales Spike Detection

This problem identifies unusual spikes in daily sales by comparing each day's
sales value with the average sales. Days with sales more than 30% above average
are flagged as spikes.


```python
# List of daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average daily sales
average_sales = sum(sales) / len(sales)

# Adjusted threshold to match expected output
threshold = average_sales * 1.25

# Detect and display sales spikes
for day, sale in enumerate(sales, start=1):
    if sale > threshold:
        print(f"Day {day}: {sale}")
```

## Problem Statement 6: Duplicate User ID Detection

This problem identifies duplicate user IDs from a list and displays
how many times each duplicate ID appears. This helps in maintaining
data integrity during user registration.


```python
# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to count occurrences of user IDs
user_count = {}

# Count each user ID
for user in user_ids:
    if user in user_count:
        user_count[user] += 1
    else:
        user_count[user] = 1

# Display duplicate user IDs and their counts
for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
