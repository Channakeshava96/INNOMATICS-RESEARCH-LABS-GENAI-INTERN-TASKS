#Problem Statement 1: Smart Parking Lot Management System

```python
def smart_parking_system(capacity, vehicle_logs):
    parked_vehicles = 0

    for log in vehicle_logs:
        if log == "IN":
            parked_vehicles += 1
        elif log == "OUT" and parked_vehicles > 0:
            parked_vehicles -= 1

    print("Currently Parked Vehicles:", parked_vehicles)

    if parked_vehicles > capacity:
        print("Parking Status: Capacity Exceeded")
    else:
        print("Parking Status: Available")


# Sample Input
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(parking_capacity, vehicle_logs)
```

#Problem Statement 2: Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance_km, traffic_level, weather):
    time = distance_km * 5

    if traffic_level == "High":
        time += 10
    elif traffic_level == "Medium":
        time += 5

    if weather == "Rainy":
        time += 5
    elif weather == "Stormy":
        time += 15

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
distance = 8
traffic = "High"
weather_condition = "Rainy"

estimate_delivery_time(distance, traffic, weather_condition)
```

#Problem Statement 3: Movie Theatre Seat Occupancy Analyzer

```python
def analyze_seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy_percentage = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy_percentage), "%")

    if occupancy_percentage == 100:
        print("Show Status: Housefull")
    elif occupancy_percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

analyze_seat_occupancy(total_seats, booked_seats)
```

#Problem Statement 4: Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", int(avg_cpu), "%")

    if avg_cpu < 50:
        print("Server Status: Normal")
    elif avg_cpu < 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

#Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def monitor_resource_usage(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resource_usage(resource_usage)
```

#Problem Statement 6: Online Event Registration Capacity Controller


```python
def manage_event_registrations(capacity, registrations):
    if registrations <= capacity:
        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users:", 0)
        print("Registration Status: Open")
    else:
        print("Confirmed Registrations:", capacity)
        print("Waitlisted Users:", registrations - capacity)
        print("Registration Status: Closed")


# Sample Input
event_capacity = 100
total_registrations = 105

manage_event_registrations(event_capacity, total_registrations)
```
