#Problem 1: Social Media – Post Engagement Analyzer

```python
# Function to analyze post engagement
def analyze_post_engagement(likes_list):
    """
    This function calculates total likes of a social media post
    and determines whether it is viral or normal.
    """
    total_likes = 0

    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes

    # Conditional check for post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# Sample data (likes received at different times)
likes_data = [200, 150, 300, 250, 150]

# Function call
total, post_status = analyze_post_engagement(likes_data)

# Display output
print("Total Likes:", total)
print("Post Status:", post_status)
```

#Problem 2: Healthcare – Medicine Stock Alert System

#Problem 3: Agriculture – Rainfall Adequacy Checker

```python
# Function to check rainfall adequacy
def check_rainfall(rainfall_list, required_level):
    """
    This function calculates the average rainfall
    and determines whether it is adequate or not.
    """
    total_rainfall = 0

    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total_rainfall += rain

    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_list)

    # Conditional check for rainfall status
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average_rainfall, status


# Sample rainfall data
seasonal_rainfall = [70, 75, 68, 74, 73]
required_rainfall_level = 70

# Function call
avg_rainfall, rainfall_status = check_rainfall(seasonal_rainfall, required_rainfall_level)

# Display output
print("Average Rainfall:", int(avg_rainfall))
print("Rainfall Status:", rainfall_status)
```

#Problem 4: Social Media – Duplicate Account Detection

```python
# Function to detect duplicate usernames
def detect_duplicate_accounts(usernames):
    """
    This function checks for duplicate usernames
    using a set data structure.
    """
    unique_usernames = set(usernames)

    # Check if duplicates exist
    if len(unique_usernames) != len(usernames):
        return "Yes"
    else:
        return "No"


# Sample username data
username_list = ["user1", "user2", "user3", "user1", "user4"]

# Function call
duplicate_status = detect_duplicate_accounts(username_list)

# Display output
print("Duplicate Accounts Found:", duplicate_status)
```

#Problem 6: Agriculture – Premium Crop Price Filter

```python
# Function to filter premium crop prices
def filter_premium_crops(price_list):
    """
    This function filters and returns crop prices
    that are considered premium.
    """
    premium_crops = []

    # Loop to filter premium crop prices
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)

    return premium_crops


# Sample crop price data
crop_prices = [1500, 2500, 1800, 3200, 2000]

# Function call
premium_crop_list = filter_premium_crops(crop_prices)

# Display output
print("Premium Crops:", premium_crop_list)
```

#Problem 7: System Monitoring – Application Health Checker

```python
# Function to check application health
def check_application_health(error_count):
    """
    This function evaluates application health
    based on the number of errors.
    """
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return error_count, status


# Sample error count
app_error_count = 7

# Function call
errors, system_status = check_application_health(app_error_count)

# Display output
print("Error Count:", errors)
print("System Status:", system_status)
```

#Problem 8: Banking – Daily Transaction Limit Checker

```python
# Function to check daily transaction limit
def check_transaction_limit(amount):
    """
    This function checks whether a transaction
    amount is within the allowed daily limit.
    """
    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    return amount, status


# Sample transaction amount
transaction_amount = 60000

# Function call
amount, transaction_status = check_transaction_limit(transaction_amount)

# Display output
print("Transaction Amount:", amount)
print("Transaction Status:", transaction_status)
```

#Problem 9: E-Learning – Student Attendance Eligibility System

```python
# Function to check attendance eligibility
def check_attendance_eligibility(attendance_list):
    """
    This function calculates attendance percentage
    and determines exam eligibility.
    """
    total_days = len(attendance_list)
    present_days = 0

    # Loop to count present days
    for day in attendance_list:
        if day == 1:
            present_days += 1

    # Calculate attendance percentage
    attendance_percentage = (present_days / total_days) * 100

    # Check eligibility
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return attendance_percentage, status


# Sample attendance data (1 = Present, 0 = Absent)
attendance_data = [1, 1, 1, 0, 1]

# Function call
percentage, eligibility = check_attendance_eligibility(attendance_data)

# Display output
print("Attendance Percentage:", percentage)
print("Exam Eligibility:", eligibility)
```

#Problem 10: Smart Electricity Bill Analyzer

```python
# Function to calculate electricity bill and usage status
def calculate_electricity_bill(units):
    """
    This function calculates the electricity bill
    based on slab rates and classifies usage level.
    """
    bill = 0

    # Slab-wise bill calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill < 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Sample units consumed
units_consumed = 250

# Function call
total_bill, usage_status = calculate_electricity_bill(units_consumed)

# Display output
print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Status:", usage_status)
```
