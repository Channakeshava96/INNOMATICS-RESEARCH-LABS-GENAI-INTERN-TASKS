PROGRAM 1 USER LOGIN CHECK 

```python
# Predefined credentials
username = "admin"
password = "1234"

# Taking input from user for username and their passkey to verify 
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Checking credential using comparison operator
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

PROGRAM 2 PASS FAIL ANALYZER

```python
# List of student marks
marks = [45, 78, 90, 33, 60]
#counter variables to keep a track of number of students who have passed or failed
pass_count = 0
fail_count = 0

# Loop through marks and counting number of students who have passed and failed
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Display results
print("Total Students:", len(marks))
print("Number of Pass Students:", pass_count)
print("Number of Fail Students:", fail_count)
```

PROGRAMM 3 DATA CLEANER

```python
# Original list given in the question 
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Cleaning process by using strip() function to clear off the additional space and lower() to change to lower case letters
for name in names:
    cleaned_name = name.strip().lower()  # remove spaces and convert to lowercase
    cleaned_names.append(cleaned_name)

# Print cleaned list
print("Cleaned Names List:", cleaned_names)
```

PEOGRAMM 4 MESSAGE LENGTH ANALYZER 

```python
#LIST of messages given in the question
messages = ["Hi", "Welcome to the platform", "OK"]
#traversing in the list to calculate lenght of each message in the list using len() function 
for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)
    
    if length > 10:
        print("ALERT!!! This message is longer than 10 characters.")
    
    print()  # blank line for better readability
```

PROGRAM 5 ERROR MESSAGE DTECTOR

```python
#word log given in the question 
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
#counter variable to count the number of 'ERROR' in the given list
error_count = 0

# Count ERROR entries my traversing every word in the log
for log in logs:
    if log == "ERROR":
        error_count += 1

# Display result
print("Total ERROR entries:", error_count)
```
