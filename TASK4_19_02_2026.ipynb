# Problem 1: Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        # Check minimum recharge amount
        if amount < 50:
            print("Invalid recharge! Amount must be at least ₹50. Please try again.\n")
            continue

        # Check for valid recharge plan
        if amount in valid_plans:
            print(f"Recharge of ₹{amount} successful ✅")
            break
        else:
            print("Invalid plan selected.")
            print("Available plans are: 199, 299, 399, 599.\n")

# Function call
validate_recharge()
validate_recharge()
validate_recharge()
```

# Problem 2: Inventory Reorder Alert System

```python
def inventory_reorder_alert(products):
    # Loop through each product and its stock quantity
    for product, stock in products.items():

        # Check if stock is below reorder level
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock: {stock})")
        else:
            print(f"{product}: Stock OK (Stock: {stock})")

# Sample product stock dictionary
products_stock = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 30
}

# Function call
inventory_reorder_alert(products_stock)
```

#Problem 3: Student Result Processing System

```python
def student_result(marks):
    total = 0

    # Calculate total marks using loop
    for mark in marks:
        total += mark

    # Calculate average marks
    average = total / len(marks)

    # Determine result
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"

    return average, result

# Sample input
student_marks = [60, 55, 70, 45, 50]

# Function call
avg, res = student_result(student_marks)

print("Average Marks:", avg)
print("Result:", res)
```

#Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak_hour):
    base_fare = 50
    per_km_rate = 12

    # Calculate basic fare
    total_fare = base_fare + (distance * per_km_rate)

    # Apply peak hour charge
    if peak_hour == "yes":
        total_fare += total_fare * 0.25

    return total_fare


while True:
    distance = float(input("Enter distance travelled (in km): "))
    peak_hour = input("Is it peak hour? (yes/no): ").lower()

    fare = calculate_fare(distance, peak_hour)
    print("Total Cab Fare: ₹", fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

#Problem 5: Employee Attendance Eligibility Checker

```python
def check_attendance(attendance):
    present_days = 0
    total_days = len(attendance)

    # Count present days using loop
    for day in attendance:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    attendance_percentage = (present_days / total_days) * 100

    # Check eligibility
    if attendance_percentage >= 75:
        return attendance_percentage, "Eligible"
    else:
        return attendance_percentage, "Not Eligible"


# Sample input
attendance_list = ["P", "P", "A", "P", "P", "P", "A", "P"]

# Function call
percentage, status = check_attendance(attendance_list)

print("Attendance Percentage:", percentage)
print("Status:", status)
```

#Problem 6: Password Strength Checker

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Loop through each character in password
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    # Check all conditions
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Sample input
password_input = "Pass@123"

# Function call
result = check_password_strength(password_input)

print("Password:", password_input)
print("Strength:", result)
```
